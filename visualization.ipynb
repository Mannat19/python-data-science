```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as py
import plotly.graph_objects as go
```

```python
df=sns.load_dataset('iris')
df
```

```python
# bar plot for sepal length
sns.barplot(x='species',y='sepal_length',data=df)
```

```python
sns.barplot(df,x='species',y='sepal_length')
```

```python
sns.barplot(df,x='species',y='sepal_width')
```

```python
sns.barplot(df,x='species',y='petal_length')
```

```python
sns.barplot(df,x='species',y='petal_width')
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(10,10),nrows=2,ncols=2)
_=sns.barplot(df,x='species',y='sepal_length',ax=ax[0,0])
_=sns.barplot(df,x='species',y='petal_length',ax=ax[0,1])
_=sns.barplot(df,x='species',y='petal_width',ax=ax[1,0])
_=sns.barplot(df,x='species',y='sepal_width',ax=ax[1,1])
```

```python
fig,ax=plt.subplots(figsize=(20,5),ncols=4)
_=sns.barplot(df,x='species',y='sepal_length',ax=ax[0])
_=sns.barplot(df,x='species',y='petal_length',ax=ax[1])
_=sns.barplot(df,x='species',y='petal_width',ax=ax[2])
_=sns.barplot(df,x='species',y='sepal_width',ax=ax[3])
```
